```python
# Importing important libraries to carry out the operations
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
%matplotlib inline
```

```python
# Dropping duplicates values so they don't interfere with our analysis
data.drop_duplicates(inplace=True)
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Fare'] = pd.to_numeric(data['Fare'], errors='coerce')
```

```python
# Histogram plot for count of different age 
sns.histplot(data['Age'].dropna(),kde=True)
```

```python
# Scatterplot for different amount of fare paid by male and female passengers 
sns.scatterplot(data=data,x=data['Age'].dropna(),y=data['Fare'].dropna(),hue='Sex')
```

```python
# Countplot of number of male and female passengers 
sns.countplot(data=data,x=data['Sex'].dropna())
```

```python
# Countplot for number of passengers in different Pclass
sns.countplot(data=data,x=data['Pclass'].dropna())
```

```python
# Distplot for density of the Fare
sns.displot(data=data, x="Fare", kind="kde")
```

```python
# Swarmplot of different Fare paid by the different sex 
sns.swarmplot(x="Sex", y="Fare", data=data,hue="Sex", palette='viridis')
```

```python
# Pairplot of all the attributes of the dataset
sns.pairplot(data)
```

```python
# Violinplot for different ages of people in different Pclass
sns.violinplot(x="Pclass", y="Age", data=data)
```

```python
# Boxplot between the fare and the different genders
sns.boxplot(x='Sex', y='Fare', data=data)
```
